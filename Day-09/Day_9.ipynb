```python
#continue

count = 0

while count < 10:
  count += 1
  if count == 3:
    continue
  print(count)
print("loop ended")
```

```python
#pass

count = 0

while count < 10:
  count += 1
  if count == 3:
    pass
  print(count)
print("loop ended")
```

```python
#number guessing game

num = 20
while True:
  number = int(input("Enter a number"))
  if(number == num):
    print("correct guess")
    break
  else:
    print("try again")
```

```python
#ATM machine

balance = 1000

while True:
    print("ATM Menu")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Your balance is:", balance)

    elif choice == 2:
        amount = int(input("Enter amount to deposit: "))
        balance = balance + amount
        print("Deposit successful!")

    elif choice < 1 or choice > 3:
        print("Invalid choice! Please try again.")

    else:
        print("Exit")
        break

```
