```python
#Q1 Write a function total_salary(*args) that accepts an employee's salary components(basic salary, allowance, bonus, etc.) and returns the total salary.

def total_salary(*args):
    total = sum(args)
    return total

print("Total Salary:", total_salary(25000, 5000, 3000, 2000))
```

```python
#Q2 Create a function cart_total(*args) that accepts the prices of any number ofproducts and returns the total bill amount.

def cart_total(*args):
    total = sum(args)
    return total

print("Total Bill Amount:", cart_total(100, 200, 300, 400, 500))
```

```python
#Q3 Write a function student_details(**kwargs) that accepts a student's information askeyword arguments and displays all the details.

def student_details(**kwargs):
  print("Student Details:")
  for key, value in kwargs.items():
      print(f"{key}: {value}")


student_details(name="Relish", age=18, faculty="DSML",roll_no=10)

```

```python
#Q4 Create a function product info(**kwargs) that accepts product details such as name, price, brand, and category, then displays them in a readable format.

def product_info(**kwargs):
  print("Prduct Details")
  for k, v in kwargs.items():
    print(f"{k}:{v}")

product_info(name = "Laptop", price = "Rs.250,000", brand = "Apple", catogery = "Electronics")
```
