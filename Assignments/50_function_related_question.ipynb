```python
#Q1 To cheak Armstrong

def arm():
  num = int(input("Enter a 3 digit number"))
  sum = 0
  org = num
  while (num > 0):
    r = num % 10
    sum = sum + r**3
    num = num // 10

  if(org == sum):
    print ("The number is armstrong",sum)
  else:
    print ("The number is not armstrong", sum)

arm()
```

```python
#Q2 Prime number between given 2 numbers

def prime():
  n1 = int(input("Enter a lower number"))
  n2 = int(input("Enter a upper number"))
  for num in range(n1, n2+1):
    if num >= 1:
      prime = True
      for i in range(2, num):
        if num % i == 0:
          prime = False
          break
      if prime:
        print(num)

prime()
```

```python
#Q5 reverse order

def rev():
  num = int(input("Enter a number"))
  rev = 0
  while (num > 0):
    r = num % 10
    rev = rev * 10 + r
    num = num // 10
  print(rev)

rev()
```

```python
#Q6 palindrome

def pal():
  num = int(input("Enter a number"))
  rev = 0
  temp = num
  while (num > 0):
    r = num % 10
    rev = rev * 10 + r
    num = num // 10
  if (temp == rev):
    print("The number is palindrome")
  else:
    print("The number is not palindrome")

pal()
```

```python
#Q8 is it perfect number

def perfect():
  num = int(input("Enter a number"))
  sum = 0
  for i in range(1, num):
      if (num % i == 0):
        sum = sum + i
  if (sum == num):
      print("The number is perfect")
  else:
      print("The number is not perfect")

perfect()
```

```python
#Q9 strong number

def strong():
  num = int(input("Enter a number: "))

  original = num
  sum = 0

  while num > 0:
    digit = num % 10

    fact = 1
    for i in range(1, digit + 1):
        fact = fact * i

    sum = sum + fact
    num = num // 10

  if sum == original:
      print(original, "is a Strong Number")
  else:
      print(original, "is not a Strong Number")

strong()
```

```python
#Q10 automorphic number

def is_automorphic(num):
    square = num * num

    if str(square).endswith(str(num)):
        print(num, "is an Automorphic Number")
    else:
        print(num, "is not an Automorphic Number")


n = int(input("Enter a number: "))
is_automorphic(n)
```

```python
#Q49 Write a function to calculate the total bill after applying a discount.

def final_bill(price, discount_percent):
    discount = (price * discount_percent) / 100
    bill = price - discount

    print("Discount Amount =", discount)
    print("Final Bill =", bill)

price = float(input("Enter Product Price: "))
discount_percent = float(input("Enter Discount Percentage: "))

final_bill(price, discount_percent)
```
