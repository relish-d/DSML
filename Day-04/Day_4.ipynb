```python
#Q2

students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 88,
    "David": 95
}

top_student = max(students, key=students.get)

print("Student with the highest marks:", top_student)
print("Marks:", students[top_student])
```
