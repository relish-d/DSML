```python
def sum(*num):
  result = 0
  for n in num:
    result += n
  print(result)

sum(1,2,3,4,5)
```

```python
def func(**kwargs):
    for k, val in kwargs.items():
        print(k, "=", val)

func(s1 = 'python', s2 = 'is', s3 = 'awesome')
```

```python
def introduce(**kwargs):
  details = []
  for k, v in kwargs.items():
    print(k,v)
    details.append(k + ":" + str(v))
  return ",".join(details)
```

```python
introduce(Name="Alice", Age="25", City="New York")
```

```python
# subject and student detials
# position argument and details keyword argument

def student_info(*args, **kwargs):
  print("Subject:", args)
  print("Details:", kwargs)
```

```python
student_info("math","science","physics",name="Dhiraj",class_="skillshikshya")
```

```python
#Write a function that accepts products and their prices as keyword arguments and returns the total.

def total_price(**products):
    total = 0
    for price in products.values():
        total += price
    return total

result = total_price(apple=120, banana=80, milk=150, bread=60)
print("Total Price =", result)
```
