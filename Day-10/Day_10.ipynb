```python
def calc(num1, num2, op):
  if op == "+":
    return num1 + num2
  elif op == "-":
    return num1 - num2
  elif op == "*":
    return num1 * num2
  elif op == "/":
    return num1/num2
  else:
    return "error symbol"
```

```python
first = float(input("enter first number:"))
second = float(input("enter second number:"))
operator = input("enter operator:")
result = calc(first, second, operator)
print("result:", result)
```

```python
def lst_operation():
  number = []
  for i in range(5):
    value = input(f"Enter value {i +1}:")
    if "." in value:
      if len(number) > 0:
        removed = number.pop()
        print(f"Float detected. removed: {removed}")
      else:
        print("float detected, but the list is empty")
    else:
      number.append(int(value))
      print(f"Integer {value} added to the list")
    print("updated List:", number)

lst_operation()
```
