```python
#square using lambda function
def square(x):
  return x**2

square(4)
```

```python
#map (function iterable)

lst = [1,2,3,4,5,6,7,8]
evens = list(filter(lambda x: x%2==0,lst))
evens
```

```python
#sort a list of tuples using lambda function
students = [
    ("Ram",90),
    ("Shyam",80),
    ("Hari",70),
    ("Gita",60)
]

students.sort(key=lambda x:x[1])

students
```

```python
#using reduce() : sum of all numbers --> lambda

#reduce(function, iterable)
from functools import reduce
numbers = [1,2,3,4,5]
result = reduce(lambda a,b:a+b, numbers)
result
```

```python
employees = [
    {"name": "Ram", "salary": 500000},
    {"name": "Hari", "salary": 30000},
    {"name":"Sita", "salary": 70000}
]

employees.sort(key=lambda emp:emp["salary"])
employees
```
