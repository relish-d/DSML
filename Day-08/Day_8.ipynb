```python
#Class task

marks = int(input("Enter your marks: "))

if marks >= 80 and marks <= 100:
    print("Grade A")
elif marks >= 60 and marks <= 79:
    print("Grade B")
elif marks >= 40 and marks <= 59:
    print("Grade C")
elif marks >= 0 and marks < 40:
    print("Grade D")
else:
    print("Invalid marks")
```
